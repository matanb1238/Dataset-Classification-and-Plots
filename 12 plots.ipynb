```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv("C:\\Users\\matan\\Downloads\\AppleStore.csv")
df = pd.DataFrame(data = d)
```

```python
# Each Category Number Of Apps
app_count = df['prime_genre'].value_counts()

app_count.plot(kind = 'bar', figsize = (15, 5), facecolor = 'r')
plt.xticks(fontsize = 12, color = 'black')
plt.yticks(fontsize = 12, color = 'black')
plt.title("Each Category Number Of Apps", fontsize = 20, color = 'r')
plt.legend(fontsize = 12)
plt.show()
```

```python
categories_mean = df.groupby('prime_genre').mean()
price_mean = categories_mean['price']
price_mean.plot(kind = 'bar', figsize = (15, 5), facecolor = 'b')
plt.xticks(fontsize = 12, color = 'firebrick')
plt.yticks(fontsize = 12, color = 'firebrick')
plt.title("Each Category Price Average", fontsize = 20, color = 'black')
plt.legend(fontsize = 12)
plt.show()
```

```python
categories_mean = df.groupby('prime_genre').mean()
lang_mean = categories_mean['lang.num']
lang_mean.plot(kind = 'bar', figsize = (15, 5), facecolor = 'lime')
plt.xticks(fontsize = 12, color = 'b')
plt.yticks(fontsize = 12, color = 'b')
plt.title("Each Category Number Of Languages Average", fontsize = 20, color = 'b')
plt.legend(fontsize = 12)
plt.show()
```

```python
categories_mean = df.groupby('prime_genre').mean()
total_rating = categories_mean['rating_count_tot']
total_rating.plot(kind = 'bar', figsize = (15, 5), facecolor = 'yellow')
plt.xticks(fontsize = 12, color = 'b')
plt.yticks(fontsize = 12, color = 'b')
plt.title("Each Category Rating Sum", fontsize = 20, color = 'black')
plt.legend(fontsize = 12)
plt.show()
```

```python

plt.subplot(1,2,1)
sns.stripplot(df.user_rating, df['sup_devices.num'], jitter=0.25, size=4, linewidth=.5)
plt.title('Sup Devices vs User Rating', fontsize=8)

plt.subplot(1,2,2)
sns.stripplot(df.user_rating_ver, df['sup_devices.num'], jitter=0.25, size=4, linewidth=.5)
plt.title('Sup Devices vs User Rating For Last Version', fontsize=8)
```

```python
#Top 15 Rated Apps
import random
sort_ratingCount = (df.sort_values(by=['rating_count_tot'], ascending = False)).head(15)
sort_ratingCount
n = sort_ratingCount['track_name'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)

plt.figure(figsize=(16,10), dpi= 80)
plt.bar(sort_ratingCount['track_name'], sort_ratingCount['rating_count_tot'], color=c, width=.5)
for i, val in enumerate(sort_ratingCount['rating_count_tot'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':50, 'size':9})

plt.gca().set_xticklabels(sort_ratingCount['track_name'], rotation=100, horizontalalignment= 'right')
plt.title("Top 15 Rated", fontsize=22)
plt.ylabel('Total Rating')
plt.show()
```
